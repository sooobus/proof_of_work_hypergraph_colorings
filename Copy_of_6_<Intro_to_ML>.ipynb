# 1. Задачи машинного обучения

Идея машинного обучения заключается в том, чтобы создать такую программу, которая для конкретной задачи принимает на вход много примеров её решения, и каким-то образом на этих примерах обучается такую задачу решать.

Есть несколько основных классов задач машинного обучения:
- Обучение с учителем (supervised learning)
- Обучение без учителя (unsupervised learning)
- Обучение с подкреплением (reinforcement learning)

Бывают и другие (например, semi-supervised learning), но так как эта область науки сейчас активно развивается, ещё нет чёткой терминологии по поводу других видов задач машинного обучения.
Мы будем в основном говорить о задачах обучения с учителем и немного о задачах обучения без учителя.
Давайте сначала разберёмся, что такое обучение с учителем и обучение без учителя.

## Обучение с учителем

Задачи обучения с учителем -- это такие задачи, для которых для некоторого набора данных известны некоторые параметры *объектов обучающей выборки* и *ответы* на эих объектах.

Примеры задач обучения с учителем: классификация и регрессия. 

**Классификация** -- это задача, где каждый объект обучающей выборки относится к какому-то классу. Задача программы -- отвечать на вопрос о принадлежности объекта к классу для новых объектов.

В задаче **регрессии** правильный ответ на каждом объекте обучающей выборки -- это число (возможно, произвольное действительное). Задача программы --- предсказывать значение этого числа на новых объектах.

## Обучение без учителя

В задачах обучения без учителя ответ на каждом элементе выборки неизвестен. Лишь известны какие-то данные, а задача алгоритма в общем случае -- в выявлении некоторых (заданных пользователем) закономерность в этих данных.

Пример задачи обучения без учителя: кластеризация.

В задаче **кластеризации** пользователь задаёт количество кластеров --- непересекающихся множеств объектов, таких, что в каждом множестве объекты в некотором смысле близки друг к другу.

## Примеры с температурой и дождём

Рассмотрим несколько примеров задач, про каждый из которых обсудим, к какому классу задач машинного обучения они относятся.

### Пример 1. Предсказание температуры

Летом Ваня каждый день выходит на балкон и измеряет температуру. Все измерения Ваня аккуратно записывает. Через 10 лет (Ваня очень упорен) получилась такая таблица температур за 5 июня:

год температура, градусы Цельсия
2000 20
2001 21
2002 20
2003 22
...
2009 16
2010 15

Ваня просыпается 4 июня на 11-ый по счёту год своих наблюдений и завтра собирается в поход. Может ли он *предсказать* на основе своих наблюдений, какая завтра будет погода?

Оказывается, что может. Как именно, узнаем чуть позже. Пока поймём, к какому классу относится задача.

Что такое элементы выборки? Это тройка (номер дня в году, номер года, температура). При этом температура -- это ответ на нашу задачу. Признаками являются номер дня в году и номер года. Кроме того, мы хотим узнать что-то про новые дни, то есть про тройки (номер дня в году, номер года, ?). Так как ответы нам на некоторых данных известны, это задача обучения с учителем. А так как ответы -- это числа, то перед нами задача *регрессии*.

### Пример 2. Будет ли дождь?

Наш герой решил в следующие 10 лет усложнить эксперимент и записывать не только температуру, но и шёл ли утром дождь. Теперь у нас есть такие записи за пятое июня:

год температура, градусы Цельсия шёл ли дождь
2000 20 да 
2001 21 нет
2002 20 нет
2003 22 нет
...
2009 16 да
2010 15 да

Теперь Ваня получает данные с высокоточного термометра, закреплённого на крыше его личного небоскрёба. Поэтому, когда он просыпается пятого июня, он уже знает температуру за окном. Но высовывать нос из-под одеяла и смотреть в окно ему не очень хочется. Можно ли предсказать на основе наблюдений, идёт ли дождь?

Что такое элементы выборки? Это четвёрки (номер дня в году, номер года, температура, шёл ли снег). Про новые дни мы знаем всё, кроме того, идёт ли снег. Значит, признаки в задаче -- номер дня в году, номер года и температура, а ответ -- идёт снег или нет. Так как выборка с известными ответами доступна, это задача обучения с учителем. Но на этот раз ответы -- не числа, а метки из конечного множества {да, нет}. Значит, перед нами задача *классификации*.

### Пример 3. Прекрасная и ужасная погода

Как вы уже поняли, Ваня довольно таки метеозависим. Один раз ему пришла в голову идея: все его дни либо ужасные, либо прекрасные, и всё в зависимости от погоды. Однако он не может точно сформулировать, какие из дней ужасные, а какие нет. Но он точно знает, что это зависит только от погоды. Ваня взял свои наблюдения (температуру и дождь) и хочет как-то разметить эти наблюдения на прекрасные и ужасные. Ваня подозревает, что ужасные дни похожи на ужасные, прекрасные на прекрасные, но прекрасные дни непохожи на ужасные (это называется *гипотеза компактности*). Может ли он это сделать?

Что такое элементы выборки? Это данные о температуре и давлении. При этом ответом является некоторая новая величина -- ужасный день или прекрасный, и эта величина неизвестна ни на одном элементе выборки. Значит, перед нами задача обучения без учителя. Ване надо разбить все дни на ужасные и прекрасные -- то есть на кластеры. Значит, перед нами кластеризация.

# 2. Признаки

Признаки -- это какие-то данные об объектах, которые нам известны. Признаки бывают нескольких типов:
-  Категориальные -- принадлежность к какому-то классу. Например, книжка по жанру бывает фантастикой или фэнтэзи
- Порядковые -- категориальные признаки, на которых задан определённый порядок. Например, отзыв о ресторане может быть плохой, средний и хороший.
